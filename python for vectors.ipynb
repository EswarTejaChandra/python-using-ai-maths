```python
# wreiien by ai                                     
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 3D vector visualization and projection to 2D
def vec3d(s1, s2, s3):
    v1 = np.array([s1, s2, s3])
    
    # 3D plot
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='r', arrow_length_ratio=0.1)
    ax.set_xlim([0, 5])
    ax.set_ylim([0, 5])
    ax.set_zlim([0, 5])
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.title('3D Vector')
    plt.show()

    # Projecting the 3D vector to 2D by ignoring the Z component (projection onto XY-plane)
    x = v1[0] + v1[1]
    y = v1[2] * 2

    print("Original 3D vector:", v1)
    print(f"Projected to 2D (x, y): ({x}, {y})")

    # 2D scatter plot
    plt.scatter(x, y, color='r')
    plt.title('Projected 2D Point')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid()
    plt.show()

# Call the function
vec3d(1, 2, 3)
```

```python
#written by ai
import numpy as np
import matplotlib.pyplot as plt

# Define a function for applying a linear transformation
def linear_transformation(matrix, vector):
    return np.dot(matrix, vector)

# Original vector (2D)
vector = np.array([4,5])

# Scaling transformation matrix (scale by 2)
scaling_matrix = np.array([[2, 0],
                           [0, 2]])

# Rotation transformation matrix (rotate by 90 degrees)
rotation_matrix = np.array([[0, -1],
                            [1, 0]])

# Apply scaling transformation
scaled_vector = linear_transformation(scaling_matrix, vector)

# Apply rotation transformation
rotated_vector = linear_transformation(rotation_matrix, vector)

# Print the results
print("Original Vector: ", vector)
print("Scaled Vector: ", scaled_vector)
print("Rotated Vector: ", rotated_vector)

# Plot the vectors for visualization
def plot_vectors(vectors, colors, labels):
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid()

    for i, vector in enumerate(vectors):
        plt.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color=colors[i], label=labels[i])

    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.legend()
    plt.show()

# Vectors to plot
vectors = [vector, scaled_vector, rotated_vector]
colors = ['blue', 'green', 'red']
labels = ['Original', 'Scaled', 'Rotated']

# Plot the vectors
plot_vectors(vectors, colors, labels)
```

```python
#ai genertaed
import numpy as np
import matplotlib.pyplot as plt

def org_vec(a, b, x, y):
    # Define vectors i and j
    i = np.array([a, b])
    j = np.array([x, y])
    
    # Projection of j onto i
    k = (np.dot(i, j) / np.dot(i, i)) * i
    
    # Plotting the vectors
    plt.quiver(0, 0, i[0], i[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector i')
    plt.quiver(0, 0, j[0], j[1], angles='xy', scale_units='xy', scale=1, color='g', label='Vector j')
    plt.quiver(0, 0, k[0], k[1], angles='xy', scale_units='xy', scale=1, color='b', label='Projection k')
    
    # Grid and labels
    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid()
    plt.legend()
    
    # Show the projection vector
    print("Vector i:", i)
    print("Vector j:", j)
    print("Projection of j onto i (k):", k)
    
    plt.show()

# Test the function
org_vec(1, 2, 2, 3)
```
